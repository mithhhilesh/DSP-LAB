```python
def fact(num):
  if(num == 0 or num == 1):
    return 1
  else:
    return num*fact(num-1)

fact(5)
```

```python
def sumOfElements(list):
  sum = 0
  for i in range(len(list)):
    sum = sum + i
  return sum
lst = [1,2,3,4,5]
sumOfElements(lst)
```

```python
def fibo(num):
  if(num == 0 or num == 1):
    return num
  else:
    return fibo(num-1) + fibo(num-2)

fibo(5)
```

```python
def calculator(string):
  return eval(string)
calculator("10-20")
```

```python
def countVowel(string):
  vowel = 0
  for i in range(len(string)):
    if(string[i] == 'a' or string[i] =='e' or string[i] =='i' or string[i] =='o' or string[i] =='u'):
      vowel = vowel + 1
  return vowel

countVowel("hey this is mithilesh i and am alive")
```
