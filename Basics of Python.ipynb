```python
def find_second_largest(numbers):
    numbers = sorted(set(numbers))
    return numbers[-2]

list1 = [10, 5, 20, 15, 20, 5]
print("The second largest element in list is:", find_second_largest(list1))

print("List without duplicates:", set(list1))
```

```python
dict1 = {'a': 10, 'b': 20, 'c': 30}
dict2 = {'b': 5, 'c': 15, 'd': 25}

merged = dict1.copy()

for key, value in dict2.items():
    merged[key] = merged.get(key, 0) + value

print(merged)
```

```python
def max_value_key(d):
    return max(d, key=d.get)

data = {'a': 10, 'b': 25, 'c': 15}
print(max_value_key(data))
```

```python
def invert_dict(d):
    inverted = {}
    for key, value in d.items():
        inverted[value] = key
    return inverted

data = {'a': 1, 'b': 2, 'c': 3}
print(invert_dict(data))
```
